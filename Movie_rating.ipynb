Importing Dependencies

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Reading the csv data file

```python
movie = pd.read_csv('/content/IMDb Movies India.csv', encoding='latin-1')
```

```python
# printing 1st 5 rows of dataframe
movie.head()
```

```python
# printing last 5 rows of dataframe
movie.tail()
```

```python
# getting shape of dataframe ( rows, columns )
movie.shape
print("No. of rows : ",movie.shape[0])
print("No. of columns : ",movie.shape[1])
```

```python
# getting info of dataframe
movie.info()
```

```python
# getting missing values in dataframe
movie.isnull().sum()
```

```python
sns.heatmap(movie.isnull())
```

```python
# dropping all missing value
movie.dropna(axis =0)
```

```python
movie.isnull().sum()
```

```python
# checking duplicate data
movie.duplicated().any()
```

```python
# describing the dataframe
movie.describe()
```

```python
# getting columns of dataframe
movie.columns
```

```python
# dropping min from duration
movie['Duration'] = movie['Duration'].str.replace(' min', '').astype(float)
```

```python
#displaying movies with more than 180 duration
movie[movie['Duration'] > 180]
```

```python
# highest average rating
movie[movie['Rating'] == movie['Rating'].max()]
```

```python
movie.columns
```

```python
# directors with average rating
movie.groupby('Director')['Rating'].mean().sort_values(ascending = False)
```

```python
# no. of movies per year
movie['Year'].value_counts()
```

```python
# plotting the grap of movies per year
sns.countplot(x = 'Year', data = movie)
```

```python
movie.columns
```

```python
# top 10 highest rating movies with directors and genres
top_10 = movie.nlargest(10, 'Rating')[['Director', 'Genre', 'Rating']]
top_10
```

```python
sns.barplot(x = 'Rating', y = top_10.index , data = top_10, hue = 'Director')
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
```

```python
# top 10 movies with highest rating with actor
top_10 = movie.nlargest(10, 'Rating')[['Actor 1','Actor 2','Actor 3', 'Rating']]
top_10
```

```python
#  average rating of movies over years
movie.groupby('Year')['Rating'].mean()
```

```python
# plotting line graph of average rating of movies over year
sns.lineplot(x = 'Year', y = 'Rating', data = movie)
```
